# Unit Conversions with Pint
In this tutorial we give the same examples as `astropy.ipynb`, but using [Pint](https://pint.readthedocs.io/) as the unit engine. This makes it easier to compare where the two packages line up and where their APIs differ.

```python
import pint

u = pint.UnitRegistry(force_ndarray=True)
u2 = pint.UnitRegistry(force_ndarray=True)
#u.formatter.default_format = "~P"
u.enable_contexts("Gaussian")
Q_ = u.Quantity
```

In Pint, a `Quantity` is also the combination of a value and a unit. The most convenient way to create a quantity is to multiply or divide a value by one of the units in the `UnitRegistry`. It works with scalars, sequences, and `numpy` arrays.

The `u.formatter.default_format = "~L"` line in the setup cell asks Pint to use compact LaTeX unit names in rich notebook display, which is closer to the Astropy output style.

```python
length = 1 * u.meter
length
```

```python
length2 = 1 * u2.meter
length2
```

```python
length*length2
```

A quantity is composed of a magnitude and units.

```python
print(length.magnitude)
print(length.units)
```

Most Pint units support prefixes, for example `k` for kilo.

```python
length = 1 * u.km
length
```

You can perform operations on quantities. Units remain and combine through the operations.

```python
time = 2 * u.s

velocity = length / time
velocity
```

Astropy has `.si` and `.cgs` convenience properties. Pint instead uses explicit conversions such as `.to_base_units()` for SI base units and `.to(...)` for the target units you want.

```python
velocity.to_base_units()
```

```python
velocity.to(u.cm / u.s)
```

It is also possible to specify explicitly which units you wish to convert to.

```python
velocity.to(u.um / u.Gs)  # micrometer per gigasecond
```

## Magnetic units

Here we show how to use magnetic units with Pint. Often in the magnetism community, a mix of CGS and SI units are used. Pint includes CGS magnetic units, and CGS/SI magnetic conversions are enabled with its `Gaussian` context.

| Quantity Symbol | SI Units | cgs Units |
|-----------------|----------|-----------|
| Length $(x)$ | $10^{-2} \, \text{m}$ | $1 \, \text{cm}$ |
| Mass $(m)$ | $10^{-3} \, \text{kg}$ | $1 \, \text{g}$ |
| Force $(F)$ | $10^{-5} \, \text{N}$ | $1 \, \text{dyne}$ |
| Energy $(E)$ | $10^{-7} \, \text{J}$ | $1 \, \text{erg}$ |
| Magnetic Induction $(B)$ | $10^{-4} \, \text{T}$ | $1 \, \text{G}$ |
| Magnetic Field Strength $(H)$ | $\frac{10^{-4}}{\mu_0} \, \text{Am}^{-1}$ | $1 \, \text{Oe}$ |
| Magnetic Moment $(\mu)$ | $10^{-3} \, \text{JT}^{-1}$ or $\text{Am}^2$ | $1 \, \text{erg G}^{-1}$ or emu |
| Magnetization $(M)$ | $10^3 \, \text{Am}^{-1}$ or $\text{JT}^{-1} \, \text{m}^{-3}$ | $1 \, \text{Oe}$ or emu cm $^{-3}$ |
| Magnetic Susceptibility $(\chi)$ | $4\pi$ (dimensionless) | $1 \, \text{emu cm}^{-3} \, \text{Oe}^{-1}$ |
| Molar Susceptibility $(\chi_m)$ | $4\pi \times 10^{-6} \, \text{m}^3 \, \text{mol}^{-1}$ | $1 \, \text{emu mol}^{-1} \, \text{Oe}^{-1}$ |
| Mass Susceptibility $(\chi_g)$ | $4\pi \times 10^{-3} \, \text{m}^3 \, \text{kg}^{-1}$ | $1 \, \text{emu g}^{-1} \, \text{Oe}^{-1}$ |
| Magnetic Flux $(\phi)$ | $10^{-8} \, \text{Tm}^2$ or $\text{Wb}$ | $1 \, \text{G cm}^2$ or $\text{Mx}$ |
| Demagnetization Factor (N) | $0 < N < 1$ | $0 < N < 4\pi$ |

```python
# Initialize quantities in cgs units
length = 1 * u.cm
mass = 1 * u.g
force = 1 * u.dyn
energy = 1 * u.erg
magnetic_induction = 1 * u.G
magnetic_field_strength = 1 * u.Oe
magnetic_moment = 1 * u.erg * u.G**-1
magnetization = 1 * u.erg / u.G / u.cm**3
magnetic_flux = 1 * u.Mx

# The Gaussian context is enabled globally, so magnetic CGS/SI conversions work directly.
print("| Quantity               | cgs Unit       | Converted SI Unit")
print("|------------------------|----------------|--------------------")
print(f"| Length                 | 1 cm           | {length.to(u.m):.2g}")
print(f"| Mass                   | 1 g            | {mass.to(u.kg):.2g}")
print(f"| Force                  | 1 dyne         | {force.to(u.N):.2g}")
print(f"| Energy                 | 1 erg          | {energy.to(u.J):.2g}")
print(f"| Magnetic Induction     | 1 G            | {magnetic_induction.to(u.T):.2g}")
print(f"| Magnetic Field Strength| 1 Oe           | {magnetic_field_strength.to(u.A / u.m):.2g}")
print(f"| Magnetic Moment        | 1 erg G^-1     | {magnetic_moment.to(u.J / u.T):.2g}")
print(f"| Magnetization          | 1 emu cm^-3    | {magnetization.to(u.A / u.m):.2g}")
print(f"| Magnetic Flux          | 1 Mx           | {magnetic_flux.to(u.Wb):.2g}")
```

There are multiple aliases for certain units. Pint includes `G` and `gauss`; unlike Astropy, `Gauss` with a capital word is not defined by default.

```python
u.G == u.gauss
```

Pint's unit definitions are documented in its default registry. For magnetic CGS/SI conversions, the important piece is the built-in `Gaussian` context.

## Contexts
Pint contexts are the closest equivalent to Astropy equivalencies. They enable conversions between quantities that are physically related but do not have the same dimensionality in the registry.

```python
exchange = 4.15 * u.K
exchange.to(u.J, "boltzmann")
```

We can also do this with magnetic units such as converting magnetic field between magnetic field strength $(\mathbf{H})$ and magnetic flux density $(\mathbf{B})$ using the relationship:

$$
\mathbf{B} = \mu_r \mu_0 \mathbf{H}.
$$

With the `Gaussian` context enabled globally, Pint can now convert `oersted` to SI magnetic flux density directly.

With the `Gaussian` context enabled globally, the free-space CGS/SI magnetic conversion works directly.

```python
H = 1 * u.Oe
H.to(u.T)
```

Pint's built-in `Gaussian` context does not take `mu_r` as an argument, so we can define a small custom context for relative permeability.

```python
def magnetic_flux_field(mu_r=1):
    context = pint.Context("magnetic_flux_field")
    source = u.oersted.dimensionality
    target = u.tesla.dimensionality

    context.add_transformation(
        source,
        target,
        lambda registry, quantity: (
            mu_r * quantity.to(u.oersted).magnitude * 1e-4 * u.tesla
        ),
    )
    context.add_transformation(
        target,
        source,
        lambda registry, quantity: (
            quantity.to(u.tesla).magnitude / (mu_r * 1e-4) * u.oersted
        ),
    )
    return context
```

```python
with u.context(magnetic_flux_field(mu_r=0.9)):
    B = H.to(u.T)
    display(B)
    display(B.to(u.G))
```

If we are doing multiple conversions, we can enable a Pint context for a block with `with ureg.context(...)`. This only applies inside the block.

```python
B = H.to(u.G)
display(B)
```

The `Gaussian` context is already enabled globally, so we can convert between SI and CGS magnetic units directly.

```python
H.to(u.T)
```

We can now convert between SI and CGS magnetic units using the enabled context.

```python
display(H)
display(H.to(u.G))
display(H.to(u.T))
display(H.to(u.A / u.m))
```

Contexts can also be used to convert other common units in magnetism.

The magnetic moment can be converted to magnetic induction by providing the volume of the respective counting unit.

In the case of formula unit:

```python
u.define("formula_unit = [formula_unit] = f_u")
u.define("atom = [atom]")
u.define("mu_B = bohr_magneton")
```

```python
import math

mu_0 = 4 * math.pi * 1e-7 * u.newton / u.ampere**2

def moment_induction(volume):
    """Context for magnetic moment per counting unit and magnetic induction."""
    if not isinstance(volume, pint.Quantity):
        raise TypeError("Volume must be a Quantity")

    volume = volume.to(u.meter**3)
    if volume.magnitude <= 0:
        raise ValueError("Volume must be positive")

    context = pint.Context("moment_induction")
    for counting_unit in [u.f_u, u.atom]:
        source = (u.mu_B / counting_unit).dimensionality
        target = u.tesla.dimensionality
        context.add_transformation(
            source,
            target,
            lambda registry, quantity, volume=volume, counting_unit=counting_unit: (
                quantity * counting_unit * mu_0 / volume
            ),
        )
        context.add_transformation(
            target,
            source,
            lambda registry, quantity, volume=volume, counting_unit=counting_unit: (
                quantity * volume / (mu_0 * counting_unit)
            ),
        )
    return context
```

```python
magnetic_moment = 8 * u.mu_B / u.f_u

vol_per_fu = (5 * u.angstrom) ** 3

with u.context(moment_induction(vol_per_fu)):
    magnetic_moment.to(u.T)
```

```python
magnetic_induction = 1 * u.T

with u.context(moment_induction(vol_per_fu)):
    magnetic_induction.to(u.mu_B / u.f_u)
```

```python
magnetic_moment = 2 * u.mu_B / u.atom

vol_per_atom = (3 * u.angstrom) ** 3

with u.context(moment_induction(vol_per_atom)):
    magnetic_moment.to(u.T)
```

```python
magnetic_induction = 1 * u.T

with u.context(moment_induction(vol_per_atom)):
    magnetic_induction.to(u.mu_B / u.atom)
```

We can also convert using other forms of induction.

```python
magnetic_induction = 10000 * u.G

with u.context(moment_induction(vol_per_atom)):
    magnetic_induction.to(u.mu_B / u.atom)
```

Pint can also stack contexts. Because the `Gaussian` context is already enabled above, this conversion can go from field strength to induction to moment in one `.to(...)`.

```python
magnetic_field_strength = 10000 * u.Oe

with u.context(moment_induction(vol_per_atom)):
    magnetic_field_strength.to(u.mu_B / u.atom)
```

## A comment on emu

emu is not a unit in the conventional manner and can have many different meanings depending on context.

Within magnetism emu can represent the magnetic moment
$$
1\, \text{emu} = 1\, \frac{\text{erg}}{\text{G}}.
$$
However, it can sometimes have dimensions
$$
1\, \text{emu} = 1\, \text{cm}^3,
$$
such as representing emu in magnetic susceptibility if defined as (emu cm $^{-3}$). Pint avoids a single built-in `emu` unit for this ambiguity; the examples above spell out `erg / gauss` or `erg / gauss / cm**3` explicitly.
